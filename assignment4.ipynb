```python
from math import comb
import matplotlib.pyplot as plt
import numpy as np

p = 0.5 
total_probability = 0  

for n in range(12, 25):  
    k = 24 - n  
    P_k = comb(n, k) * (p**k) * ((1-p)**(n-k))
    total_probability += P_k

total_probability
```

```python
p = [0] * 25
p[0] = 1

prob_toss = 0.5

for current_sum in range(0, 24):
    if current_sum + 1 <= 24:
        p[current_sum + 1] += p[current_sum] * prob_toss
    if current_sum + 2 <= 24:
        p[current_sum + 2] += p[current_sum] * prob_toss

p[24]
```

```python
max_sum = 50
probabilities = np.zeros(max_sum + 1)
probabilities[0] = 1 

for current_sum in range(max_sum):
    if current_sum + 1 <= max_sum:
        probabilities[current_sum + 1] += probabilities[current_sum] * 0.5
    if current_sum + 2 <= max_sum:
        probabilities[current_sum + 2] += probabilities[current_sum] * 0.5

plt.figure(figsize=(10, 6))
plt.plot(range(1, max_sum + 1), probabilities[1:], marker='o')
plt.title('Probability Distribution for Sums from 1 to 50')
plt.xlabel('Sum')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```
